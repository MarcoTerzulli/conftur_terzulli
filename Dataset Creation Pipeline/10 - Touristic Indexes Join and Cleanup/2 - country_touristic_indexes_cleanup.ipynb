## Dataset Import

````python
df_country_touristic_indexes = pd.read_csv(path_file_import + 'all_indexes_by_country.csv', low_memory=False, index_col=[0])
df_country_touristic_indexes
````

## Dataset Cleanup

#### Reindexing

````python
df_country_touristic_indexes = df_country_touristic_indexes.reset_index(level=0)
df_country_touristic_indexes = df_country_touristic_indexes.rename(columns={"index": "ConferenceLocation"})
df_country_touristic_indexes.drop(columns=["ConferenceLocation"], inplace = True)

df_country_touristic_indexes
````

#### Drop of the Duplicated Values

````python
df_country_touristic_indexes = df_country_touristic_indexes.drop_duplicates(subset=["Country"])
df_country_touristic_indexes
````

#### Fix of the Data Types

````python
cols= [i for i in df_country_touristic_indexes.columns if i not in ["Country"]]
for col in cols:
    df_country_touristic_indexes[col] = pd.to_numeric(df_country_touristic_indexes[col])
````

## Dataset Split

We're going to split the dataframe in two parts:
* Yearly Tourist Arrivals
* Touristic Indexes

This will allow us to use the dataset in a simpler way.

#### Drop of the Touristic Indexes Columns

Before doing this, remember to specify the year range in the following cell:

````python
start_year = 1995
final_year = 2020
````

````python
df_country_yearly_tourist_arrivals = df_country_touristic_indexes.copy()

column_to_drop = list(set(df_country_yearly_tourist_arrivals.columns.values.tolist()) - set([str(x) for x in range(start_year, final_year + 1)]) - set(["Country"]))

df_country_yearly_tourist_arrivals = df_country_yearly_tourist_arrivals.drop(column_to_drop, axis=1)
df_country_yearly_tourist_arrivals.head(3)
````

#### Drop of the Tourist Arrival Columns


````python
column_to_drop = [str(x) for x in range(start_year, final_year + 1)]

df_country_touristic_indexes = df_country_touristic_indexes.drop(column_to_drop, axis=1)
df_country_touristic_indexes.head(3)
````

## Write of the Final CSV on Disk

Saving the resulting dataframe on disk in CSV format.

````python
# Write of the resulting CSVs on Disk
df_country_touristic_indexes.to_csv(path_file_export + 'out_country_touristc_indexes.csv')
print(f'Successfully Exported the Processed CSV to {path_file_export}out_country_touristc_indexes.csv')


df_country_yearly_tourist_arrivals.to_csv(path_file_export + 'out_country_yearly_tourist_arrivals.csv')
print(f'Successfully Exported the Processed CSV to {path_file_export}out_country_yearly_tourist_arrivals.csv')
````
