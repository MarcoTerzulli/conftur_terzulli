```python
# ***************************************************************
# Preprocess OpenCitations (COCI) Dataset
# ***************************************************************

# Jupyter Notebook che si occupa del preprocessing del dataset di Opencitations (COCI)
#
# Nello specifico, vengono effettuate le seguenti operazioni:
# - apertura (sequenziale) dei CSV del Dump
# - Drop delle colonne superflue (viene mantenuta solo 'cited')
# - Gruop by sulla colonna 'cited' e conteggio
# - Unione con il dataframe dei CSV già elaborati
#
# Infine, l'intero dump preprocessato viene salvato su disco in formato CSV
```

```python
# Libraries Import
import pandas as pd
```

```python
# Dumps Directory Path
path_file_import = r'/Users/marcoterzulli/File/Scuola Local/Magistrale/Materiale Corsi Attuali/Tirocinio/Cartella di Lavoro/Archivi Dump di Lavoro/'

# CSV Exports Directory Path
#path_file_export = r'./Users/marcoterzulli/File/Scuola Local/Magistrale/Materiale Corsi Attuali/Tirocinio/Cartella di Lavoro/Repository Github/Repository Mio/terzulli_conftur/Test Pipeline Download e Preprocess/'
path_file_export = r'/Users/marcoterzulli/File/Scuola Local/Magistrale/Materiale Corsi Attuali/Tirocinio/Cartella di Lavoro/Archivi Dump di Lavoro/Export/'
```

```python
# Read CSV COCI
df_coci_current_csv = pd.read_csv(path_file_import + '2022-01-21T100308_0-4_1/2022-01-21T100308_4_1.csv')
```

```python
df_coci_current_csv
```

```python
# Drop delle colonne superflue
# Nello specifico, si droppano le colonne 'oci', 'citing', 'creation', 'timespan', 'journal_sc', 'author_sc'
df_coci_current_dropped_columns = df_coci_current_csv
df_coci_current_dropped_columns.drop(columns=['oci', 'citing', 'creation', 'timespan', 'journal_sc', 'author_sc'])
```

```python
# Raggruppamento e conteggio per articolo cited
sf_coci_current_grouped = df_coci_current_dropped_columns.groupby(['cited'])['cited'].count()

# L'oggetto restituito è di tipo Pandas Series. Conversione a Pandas dataframe
df_coci_current_grouped = pd.DataFrame({'article':sf_coci_current_grouped.index, 'citations_count':sf_coci_current_grouped.values})
```

```python
df_coci_current_grouped
```

```python
# ************************************
# Lettura secondo CSV di Test
```

```python
# Read Second CSV COCI
df_coci_current_csv_2 = pd.read_csv(path_file_import + '2022-01-21T100308_0-4_1/2022-01-21T100308_3_1.csv')
```

```python
# Drop delle colonne superflue
# Nello specifico, si droppano le colonne 'oci', 'citing', 'creation', 'timespan', 'journal_sc', 'author_sc'
df_coci_current_dropped_columns_2 = df_coci_current_csv_2
df_coci_current_dropped_columns_2.drop(columns=['oci', 'citing', 'creation', 'timespan', 'journal_sc', 'author_sc'])
```

```python
# Raggruppamento e conteggio per articolo cited
sf_coci_current_grouped_2 = df_coci_current_dropped_columns_2.groupby(['cited'])['cited'].count()

# L'oggetto restituito è di tipo Pandas Series. Conversione a Pandas dataframe
df_coci_current_grouped_2 = pd.DataFrame({'article':sf_coci_current_grouped_2.index, 'citations_count':sf_coci_current_grouped_2.values})
```

```python
# Concatenazione df elaborati
df_coci_concat = pd.concat([df_coci_current_grouped, df_coci_current_grouped_2])
```

```python
# Nuovo Raggruppamento e conteggio per articolo cited sui df concatenati
sf_coci_concat_grouped = df_coci_concat.groupby(['article'])['article'].count()

# L'oggetto restituito è di tipo Pandas Series. Conversione a Pandas dataframe
df_coci_concat_grouped = pd.DataFrame({'article':sf_coci_concat_grouped.index, 'citations_count':sf_coci_concat_grouped.values})
```

```python
df_coci_concat_grouped
```

```python
# Export on CSV
df_coci_concat_grouped.to_csv(path_file_export + 'out_coci_citations_count.csv')
```
